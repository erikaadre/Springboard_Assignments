### QUESTION 3
Write an SQL query to show a list of facilities that charge a fee to members, 
where the fee is less than 20% of the facility's monthly maintenance cost.
Return the facid, facility name, member cost, and monthly maintenance of the
facilities in question.

```python
qry = """
SELECT 
    facid,
    name as 'Facility Name',
    membercost, 
    monthlymaintenance 
FROM Facilities AS F
WHERE F.membercost < (0.2 * F.monthlymaintenance)"""


df3 = pd.read_sql_query(qry,connect)
df3
```

### QUESTION 5
Produce a list of facilities, with each labelled as
'cheap' or 'expensive', depending on if their monthly maintenance cost is
more than $100. Return the name and monthly maintenance of the facilities
in question.

```python
qry = """
SELECT 
    name,
    CASE WHEN monthlymaintenance > 100 THEN 'expensive'
    ELSE 'cheap' END AS Label
FROM Facilities 
"""


df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 7
Produce a list of all members who have used a tennis court. Include in your output the name of the court, and the name of the member formatted as a single column. Ensure no duplicate data, and order by the member name.

```python
qry = """
SELECT 
   DISTINCT m.firstname || ' ' || m.surname AS FullName,
   f.name as CourtName
FROM Members AS m
LEFT JOIN Bookings AS b
    ON m.memid = b.memid
LEFT JOIN Facilities AS f
    ON b.facid = f.facid
WHERE f.name LIKE '%Tennis Court%'
ORDER BY FullName"""


df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 8
Produce a list of bookings on the day of 2012-09-14 which will cost the member (or guest) more than $30. Remember that guests have different costs to members (the listed costs are per half-hour 'slot'), and the guest user's ID is always 0. Include in your output the name of the facility, the name of the member formatted as a single column, and the cost. Order by descending cost, and do not use any subqueries.

```python
qry = """
SELECT
    f.name AS FacilityName,
    m.firstname || ' ' || m.surname AS FullName,
    CASE WHEN m.memid = 0 THEN (b.slots * f.guestcost)
    ELSE (f.membercost*b.slots) END AS Cost
FROM Bookings as b
LEFT JOIN Members as m
    ON b.memid = m.memid
LEFT JOIN Facilities as f
    ON b.facid = f.facid
WHERE date(b.starttime) = '2012-09-14' AND
    (CASE WHEN m.memid = 0 THEN (b.slots * f.guestcost) 
    ELSE (f.membercost*b.slots) END) > 30
ORDER BY Cost DESC
    """


df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 9
This time, produce the same result as in Q8, but using a subquery.

```python
qry = '''
SELECT 
    subq.FacilityName,
    m.firstname || ' ' || m.surname AS FullName,
    subq.cost
FROM Members as m
INNER JOIN
(SELECT 
  f.name as FacilityName,
  b.memid as MemberID,
  CASE WHEN b.memid = 0 THEN (b.slots * f.guestcost)
  ELSE (f.membercost*b.slots) END AS Cost   
FROM Bookings as b
LEFT JOIN Facilities as f
ON b.facid = f.facid
WHERE date(b.starttime) = '2012-09-14') as subq
ON m.memid = subq.MemberID
WHERE subq.cost > 30
ORDER BY Cost DESC
'''


df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 10
Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
qry = '''
SELECT 
  f.name as FacilityName,
  SUM(CASE WHEN b.memid = 0 THEN (b.slots * f.guestcost)
  ELSE (f.membercost*b.slots) END) AS Revenue   
FROM Bookings as b
LEFT JOIN Facilities as f
ON b.facid = f.facid
GROUP BY FacilityName
HAVING Revenue < 1000
ORDER BY Revenue
'''


df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 11
Produce a report of members and who recommended them in alphabetic surname,firstname order 

```python
qry = '''
WITH rfltbl AS
(SELECT surname ||', ' || firstname AS Member, memid 
 FROM Members)
 
SELECT r.Member, m.surname ||', ' || m.firstname AS Referral
FROM rfltbl as r
LEFT JOIN Members as m
ON r.memid = m.recommendedby
WHERE m.recommendedby != ''
ORDER by r.Member
'''


df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 12
Find the facilities with their usage by member, but not guests

```python
qry = """
SELECT f.name, 
       SUM(CASE WHEN memid != 0 THEN 1
           ELSE 0 END) AS UsageCount_Mem
FROM Bookings b
INNER JOIN Facilities f
    ON b.facid = f.facid
GROUP BY f.name
ORDER BY UsageCount_Mem DESC
            """

df = pd.read_sql_query(qry,connect)
df
```

### QUESTION 13
Find the facilities usage by month, but not guests

```python
qry = """
SELECT f.name, 
       strftime('%m', b.starttime) as Month,
       SUM(CASE WHEN b.memid != 0 THEN 1 ELSE 0 END) AS MemberUsage
FROM Bookings AS b
INNER JOIN Facilities AS f
    ON b.facid = f.facid
GROUP BY f.name, Month
ORDER BY MemberUsage DESC
            """

df = pd.read_sql_query(qry,connect)
df
```
